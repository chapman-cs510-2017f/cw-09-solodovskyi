## Problem 5

1. Makefile directs make  on how to compile and link a program. Using C/C++ as an example, when a C/C++ source file is changed, it must be recompiled. If a header file has changed, each C/C++ source file that includes the header file must be recompiled to be safe.Basically it consists of “rules” in the following form like target, dependencies system command(s) A target is usually the name of a file that is generated by a program.

2. Initially, In the code we  have the source file and header file, and the header file can also be implemented in the source file instead of being a seperated file.It  will use c processor and c compiler to compile them into the object file. If it defines the dependence of other object files, c linker will link all of these object files together to create the executable file. This structure is reasonable, because in this I can very well understand the program structure without reading the long source file. Moreover, it would be hard for me if I need to go through c preprocessor, c complier, and c linker step by step without makefile. Make makes life easier.

3. The header file defines the types for functions that are used in the source file.

4. The header file got extension .h which contains C function declarations and macro definitions to be shared between several source files .A source file defines all the executable part except types.

5. An object file is a compiled file that we get from the source file and header file through c preprocessor and c complier. The code in an object file is machine code.

6. Compiling creates object files that consist of machine code.

7. Linking is combining the object files together to create the executable file. It defines the dependence of object files in the makefile.
